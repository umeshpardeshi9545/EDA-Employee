# Data Pre-Processing

```python
#Remove Duplicated value
df.drop_duplicates(inplace=True)
```

```python
# After removing duplicates
df.duplicated().sum()
```

```python
#Outliers treatments
def outler_treat(df,col):
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  lower_extream=q1-1.5*iqr
  upper_extream=q3+1.5*iqr
  return lower_extream,upper_extream
```

```python
#outlier treatmen on age
lower_extream,upper_extream= outler_treat(df,'age')
df['age']=np.where(df['age']>upper_extream,upper_extream,df['age'])
df['age']=np.where(df['age']<lower_extream,lower_extream,df['age'])
```

```python
sns.boxplot(df['age'])
```

```python
#outlier treatmen on fnlwgt
lower_extream,upper_extream= outler_treat(df,'fnlwgt')
df['fnlwgt']=np.where(df['fnlwgt']>upper_extream,upper_extream,df['fnlwgt'])
df['fnlwgt']=np.where(df['fnlwgt']<lower_extream,lower_extream,df['fnlwgt'])
```

```python
sns.boxplot(df['fnlwgt'])
```

```python
#outlier treatmen on education_num
lower_extream,upper_extream= outler_treat(df,'education_num')
df['education_num']=np.where(df['education_num']>upper_extream,upper_extream,df['education_num'])
df['education_num']=np.where(df['education_num']<lower_extream,lower_extream,df['education_num'])
```

```python
#outlier treatmen on capital_gain
lower_extream,upper_extream= outler_treat(df,'capital_gain')
df['capital_gain']=np.where(df['capital_gain']>upper_extream,upper_extream,df['capital_gain'])
df['capital_gain']=np.where(df['capital_gain']<lower_extream,lower_extream,df['capital_gain'])
```

```python
categorical_col5 = df[['sex','income']]
categorical_col5
```

```python
encoded_categorical_col5 = pd.get_dummies(categorical_col5, columns=['sex', 'income'], prefix=['sex', 'income'])
encoded_categorical_col5.head()
```

```python
# Checking Column with more than 5 unique value
for i in df.columns.tolist():
  if df[i].nunique() >=5 and df[i].dtype=='object':
   print("No. of unique values in ",i, "is",df[i].nunique() ,".")
```

```python
categorical_col_5plus = df[['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country']]
categorical_col_5plus
```

```python
# Label Encoder
from sklearn.preprocessing import LabelEncoder



# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to each column
for col in categorical_col_5plus.columns:
    categorical_col_5plus[col] = label_encoder.fit_transform(categorical_col_5plus[col])
```

```python
categorical_col_5plus
```

One-Hot Encoding and Label Encoding

1. One-Hot Encoding
* Its convert categorical column into numerical columns.
* In one hot coding all the value replace with 0 and 1.
* Its leads to an issue like complex model building because after One-Hot Encoding dimension of data set increase

2. Label Encoding
* Label Encoding also conver categorical column into numerical columns.
* Label Encoding flows ordinal value and replace with unique numer as per order
eg we have column contain this uniue value  (good,better,best) afte Label Encoding this value replce with (1,2,3) respectively
* After Label Encoding its very hard to iterpet as compared to orignal value.

## Final data set after normalize and encoding

```python
final_df = pd.concat([scaled_df, encoded_categorical_col5, categorical_col_5plus], axis=1)

final_df.head()
```

#Feature Engineering:

```python
# creating total capital column using net gain and loss
final_df['total_capital'] = final_df['capital_gain'] - final_df['capital_loss']
```

```python
final_df.head()
```

```python
# creating gain per hour column using total gain

final_df['gain_per_hour'] = final_df['total_capital'] / final_df['hours_per_week'].replace(0, 1)
```

```python
df.head()
```

```python
# As i have crated new 2 column which is total_capital and 	gain_per_hour so applying log transformation on this columns
final_df['log_gain_per_hour'] = np.log1p(final_df['gain_per_hour'])
final_df['log_total_gain'] = np.log1p(final_df['total_capital'])
```

# Feature Selection:

```python
pps_matrix = pps.matrix(final_df)
```

```python
# Display PPS matrix
print(pps_matrix[['x', 'y', 'ppscore']].head())
```

```python
pps_heatmap = pps_matrix.pivot(index='x', columns='y', values='ppscore')
sns.heatmap(pps_heatmap, annot=True, cmap='coolwarm', cbar=True)
plt.title("Predictive Power Score (PPS) Heatmap")
plt.show()
```

```python
final_df.corr(numeric_only=True)
plt.figure(figsize=(14, 6))
sns.heatmap(final_df.corr(numeric_only=True),annot=True)
```

* PPS may draw attention to relationships—like category and numerical feature dependencies—that the correlation matrix overlooks.
* PPS makes feature engineering more actionable by calculating how well a feature predicts another.
* By emphasizing multicollinearity, which PPS doesn't directly address, the correlation matrix enhances PPS.
* PPS value ranges between 0 to 1, While in Correlation value Ranges from -1 to * PPS capture non linear relationship, other hand correlation captute only linera realtionship.
